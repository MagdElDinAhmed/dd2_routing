```python
# Function to read file content
def read_file(file_path):
    with open(file_path, "r") as file:
        return file.read()
```

```python
# File paths
input_file = "input.txt"
output_file = "output.txt"

# Read files
input_data = read_file(input_file)
output_data = read_file(output_file)
```
